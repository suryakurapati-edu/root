```python
# Plot the distribution of variables such as Sepal length and width, Petal length and width
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
sns.histplot(data['sepal_length'], kde=True)
plt.title('Distribution of Sepal Length')

plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 2)
sns.histplot(data['sepal_width'], kde=True)
plt.title('Distribution of Sepal Width')

plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 3)
# plt.xlim(0, 350)
sns.histplot(data['petal_length'], kde=True)
plt.title('Distribution of Petal Length')

# Plot the distribution of variable Total ranging only between 0-50 to get a better understanding
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 4)
sns.histplot(data['petal_width'], kde=True)
plt.title('Distribution of Petal Width')

plt.tight_layout()
plt.show()
```

Inference:
- Plot 1 is the "Distribution of Sepal Length" plot, where the length of sepal varies between ~4-8 with most species lying between 5 - 6.25 length
- Plot 2 is the "Distribution of Sepal Width" plot, where it clearly shows a normal distribution with mean and median around 3 cm width
- Plot 3 is the "Distribution of Petal Length" plot, where it clearly shows that length of a petal between 1 - 1.7 has the highest occurence. However, lengths between 4.2 - 6.2 have constant high occurance, hence mean would lie between 3.5-5.5 although 1-1.7 has the highest occurence
- Plot 4 is the "Distribution of Petal Width" plot, where it clearly shows that width of a petal between 0.1 - 0.4 has the highest occurence. However, follows the same pattern of petal lenght with constant high occurance shown between 1.2 - 2.5, hence the mean would be between range of 1.2 - 2.5

```python
# Plot the time series of Sepal length and width, Petal length and width
plt.figure(figsize=(14, 7))
plt.plot(data['species'], data['sepal_length'], label='Sepal Length', linestyle="--", linewidth=1)
plt.plot(data['species'], data['sepal_width'], label='Sepal Width', linestyle="--", linewidth=1)
plt.plot(data['species'], data['petal_length'], label='Petal Length', linestyle="--", linewidth=1)
plt.plot(data['species'], data['petal_width'], label='Petal Width', linestyle="--", linewidth=1)
plt.xlabel('Species')
plt.ylabel('Values')
plt.title('Line Plot of Sepal and Petal Length and Width') 
plt.legend()
plt.grid(True)
plt.show()
```

Infernece:
A line plot is drawn between 4 numerical variables (sepal length, width and petal length, width) on y-axis and Iris flower 3 species i.e. setosa, versicolor and virginica on x-axis
