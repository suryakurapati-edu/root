```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('Sample_data.csv')
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
data.columns
```

```python
data['Total Price'] = data['qty'] * data['unit_price']
data.head()
```

```python
data['datetime'] = data['date'] + ' ' + data['time']
data.head()
```

```python
data['date'] = data['date'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y'))
data['datetime'] = data['datetime'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y %H:%M:%S'))
data['year'] = data['date'].apply(lambda x: x.year)
data['month'] = data['date'].apply(lambda x: x.month)
data['hour'] = data['datetime'].apply(lambda x: x.hour)
data.head()
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.drop(['id', 'product_id'], axis=1, inplace=True)
```

```python
# Comparison of sales by month
%matplotlib inline
plt.figure(figsize=(5, 3))
data.groupby('month')['Total Price'].sum().plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')
plt.show()
```

```python
# Total Sales by Category
plt.figure(figsize=(5,3))
data.groupby('category')['Total Price'].sum().plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Category')
plt.show()
```

```python
# Total sales by category and month
sales_by_category_month = pd.pivot_table(data=data, index='category', columns='month', values='Total Price', aggfunc='sum')

plt.figure(figsize=(5,10)) 
sales_by_category_month.plot(kind='bar', stacked=True)
plt.show()
```

```python
# distribution of sales by hour
plt.figure(figsize=(10,3))
data.groupby('hour')['Total Price'].sum().plot(kind='line')
plt.xticks(np.arange(0, 25, 1))
plt.grid()
plt.show()
```

```python
# Distribution of Sales
plt.figure(figsize=(5,5))
sns.boxplot(data['Total Price'])
plt.show()
```

```python
# Distribution of sales by removing outliers
plt.figure(figsize=(5,5))
sns.boxplot(data['Total Price'], showfliers=False)
plt.show()
```

```python
# Distribution of store_location by sales
plt.figure(figsize=(5,3))
data.groupby('store_loc')['Total Price'].sum().plot(kind='bar')
plt.show()
```

```python
# Relationship between sales, quantity and unit price
plt.figure(figsize=(5,3))
sns.heatmap(data[['Total Price', 'qty', 'unit_price']].corr(), annot=True, cmap='GnBu')
```

```python
df_encoded = pd.get_dummies(columns=['species'], data=df, prefix='species')
df_encoded.head()
```

```python
df_encoded['species_setosa'].value_counts()
```

```python
df_encoded['species_setosa'] = df_encoded['species_setosa'].apply(lambda x: 1 if x == True else 0)
df_encoded['species_versicolor'] = df_encoded['species_versicolor'].map({True:1, False:0})
df_encoded['species_virginica'] = df_encoded['species_virginica'].map({True:1, False:0})
df_encoded.head()
```

```python
# Feature Extraction
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species'])
df.head()
```

```python
df.head()
```

```python
df_encoded.head()
```

```python
# Feature Selection
from sklearn.ensemble import RandomForestClassifier

X = df.drop('species', axis=1)
y = df['species']

randomn_forest = RandomForestClassifier()

model = randomn_forest.fit(X,y)

importance = model.feature_importances_

for feature, score in zip(X.columns, importance):
    print(feature, score)
```
